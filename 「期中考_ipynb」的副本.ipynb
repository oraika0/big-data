另外一種圖 只有給一個軸的資料:TotalPoints

```
qplot(FieldGoalsAttempted,TotalPoints,
data=NBA1314,
facets=Position~.)
```

```
qplot(FieldGoalsAttempted,TotalPoints,
data=NBA1314,
facets=.~Position)
```

```
qplot(TotalPoints,data=NBA1314,
facets=Position~.,binwidth=10)
```

```
qplot(TotalPoints,data=NBA1314,
facets=Position~.,binwidth=100)
```
